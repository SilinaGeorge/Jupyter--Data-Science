Price, Year, EngineV and Mileage are likely the ones that will be a challenge.
Lets check for linearity using scatter plot


```python
f, (ax1,ax2,ax3) = plt.subplots(1,3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')

plt.show()
```

we can spot patterns in these graphs, but not linear ones
=> we should not run a linear regression. We should first transform one or more vars. Use log transformation b/c these scatter plots are exponential 

```python
sns.distplot(data_cleaned['Price'])
```

### Relaxing the assumptions

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned ['log_price'] = log_price
data_cleaned
```

```python
f, (ax1,ax2,ax3) = plt.subplots(1,3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('log_price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('log_price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('log_price and Mileage')

plt.show()
```

there is now a linear pattern in all plots

```python
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

no endogeneity, normailty and homoscedasticity, no autocorrelation are all true. now we need to check for multicollinearity

## Multicollinearity

```python
data_cleaned.columns.values
```

common knowledge that year and mileage is going to be correlated.
sklearn doesnt has a dedicated method to check autocorrelation are all true. now we need to check for multicollinearity so we will use statsmodels.
one of the best ways to check for multicollinearity is through **VIF (variance inflation factor)**. It produces a measure which estimates how much larger the square root of the std. error of an estimate is compared to a situation where the var. was compeltely uncorrected with the other predictors

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year','EngineV']] #define features we want to check for multi. col.
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range (variables.shape[1])]
vif['features'] = variables.columns
vif
```

vif = 1 => no multi. col.
1 < vif < 5 => perfectly ok
vif >5 or vif >6 or vif >10 (depends on who you ask) is not ok

for our example, Year is definetely too correlated with out other variables so we will only remove year

```python
data_no_mult_col = data_cleaned.drop(['Year'], axis=1)
```

 ## Create dummy variables

for categorical vars.
**pd.get_dummmies(df, drop_first)** spots all categorical vars. and creates dummies automatically

if we have N categories for a feature, we have to create N-1 dummies

```python
data_with_dummies = pd.get_dummies (data_no_mult_col, drop_first=True)
data_with_dummies.head()
```

no dummie for brand Audo is not crreated because if the rest is all 0 it is obviously an Audi. We do this to make sure we dont have mult. col. because if we did create a var for Audio it would be dependant on the other vars.

## rearrange a bit

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear Regression Model

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

### scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

not usually recommended to standarize dummy vars. However in ML we dont give it much thought into it because scaling has no effect on the predictive power of dummies. Once scaled, though, they lose their dummy meaning

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.3,random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

this is a log-linear regression as the dependent var. is the log of Price

```python
y_hat = reg.predict(x_train)
```

a simple wy o check the final result is to plot the predicted values against the observed values. obeserved values = targets

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets(y_train)', size=10)
plt.ylabel('Predictions(y_hat)', size=10)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show() # targets and predictions
```

best case senario: if the target is 7 we want the prediction to be 7. if target is 10 we want prediction to be 10 etc.

our plot is close, so its good

Another check is the residual plot which is the distribution of the residuals. 

residual = differences between the targets and predictions

```python
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF', size=18)
```

normally distributed
mean = 0
however the trail on the left side is much longer than right => there are certain obs. for which (y_train-y_hat) is much lower than the mean (a much higher price is predicted than is observed, overestimate). Predictions rarely underestimate targets 

```python
reg.score(x_train, y_train)
```

 our model is explaining 75% of the variability of the data

### Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

**WEIGHTS INTERPRETATION - Numerical**
1. A Positibe weight shows that a feature increases in value, so does the log_price and Price respectively e.g. EngineV, the bigger the engine volume the higher the price

2. a negative weight shows that as a feature increases in value, log_price and Price decreases e.g. Mileage, the more the car is driven the lower the price 

**Mileage is the most prominant feature in this regression (biggest weight)** it is almost twice as important as EngineV




```python
data_cleaned['Brand'].unique()
```

when all other dummies are zero Audi = 1 because Audi is the dropped Brand (Audi not in table above) => **Audi is the benchmark**

**WEIGHTS INTERPRETATION - Dummy Vars**

1. a positive weight shows that the respective category (Brand) is more expensive than the benchmark (Audi) e.g. on avg BMW more expensive than Audi

2. a negative weight shows that the respective category (Brand) is less expensive than the benchmark (Audi) e.g. on avg Mitsubishi less expensive than Audi

Mercedes-Benz is very close to 0 => ong avg it is very close to the price of an Audi
